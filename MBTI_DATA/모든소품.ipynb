```python
import pandas as pd
```

```python
df = pd.read_csv('소품 라벨링 8067360f081543bdb6810475e5963558.csv')
df
```

```python
del(df['Column'])
```

```python
df.dropna(inplace=True)
```

```python
df['무드']
```

```python
mood = []
for i in range(len(df.index)):
    mood.append(df['무드'][i].split(', '))
```

```python
mood
```

```python
moods=[]
for i in range(len(df.index)):
    for j in range(3):
        moods.append(mood[i][j])
```

```python
moods
```

```python
mood1 = []
mood2 = []
mood3 = []
for i in range(0,len(df.index)*3,3):
    mood1.append(moods[i])
for i in range(1,len(df.index)*3,3):
    mood2.append(moods[i])
for i in range(2,len(df.index)*3,3):
    mood3.append(moods[i])
```

```python
df['mood1'] = mood1
df['mood2'] = mood2
df['mood3'] = mood3
```

```python
del(df['무드'])
```

```python
df.to_csv('goods.csv')
```
