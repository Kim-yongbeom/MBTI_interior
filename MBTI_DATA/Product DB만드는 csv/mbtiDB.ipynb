```python
label.columns = ['name','mood','category','mood_pic','detail_pic']
label
```

```python
mood_list = []
mood = label['mood'] 
for i in range(len(mood)):
  mood_list.append(mood[i].split(', '))

mood_list
```

```python
aa = pd.DataFrame(mood_list, columns=['mood1','mood2','mood3'])
aa
```

```python
label = pd.concat([label,aa], axis=1)
label
```

```python
one = pd.get_dummies(aa['mood1'])
two = pd.get_dummies(aa['mood2'])
three = pd.get_dummies(aa['mood3'])
```

```python
two[['vintage','casual']] = 0
two
```

```python
three[['individuality','casual','minimal']] = 0
three
```

```python
three = three[['minimal','modern','natural','casual','romantic','individuality','useful','vintage']]
three
```
